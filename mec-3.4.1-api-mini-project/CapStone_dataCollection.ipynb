Retrieve data from news api

```python
import requests
import json
import os
import urllib.request
import bs4
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
with open("./key.json") as key_file:
    key = json.load(key_file)
```

```python
url = 'https://api.nytimes.com/svc/topstories/v2/us.json'
header = {"api-key":key['api_key']}
```

```python
r = requests.get(url, params=header)
dictionary = json.loads(r.text)
metadata = dictionary['results'][0]
article_metadata = dictionary['results'][1:]
```

```python
urls = [a['url'] for a in article_metadata]
```

```python
uClient = uReq(urls[0])
page_html = uClient.read()
uClient.close()
```
